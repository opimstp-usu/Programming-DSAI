```python
import matplotlib.pyplot as pyplot
import random 
```

Create a sequence of random values

```python
def create_random_array(num_arr):
    myArr = []
    for i in range(num_arr):
        myArr.append(random.random())
    return myArr
```

Plot the values

```python
my_arr = create_random_array(100)
pyplot.plot(my_arr)
```
