```python
#Importing pyplot
from matplotlib import pyplot as plt
from matplotlib import style
import random 
```

```python
def create_random_array(num_arr):
    myArr = []
    for i in range(num_arr):
        myArr.append(random.randint(1,100))
    return myArr
```

```python
x = create_random_array(10)
y = create_random_array(10)
x1 = create_random_array(10)
y1 = create_random_array(10)
```

```python
style.use('ggplot')
```

```python
plt.plot(x,y,'g', linewidth=5, label='line one')
plt.plot(x1,y1,'c', linewidth=5, label='line two')

plt.title('Random Data')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.legend()
plt.grid(True,color='k')

plt.show()
```

```python
style.use('ggplot')
plt.bar(x, y, align='center')
plt.bar(x1, y1, color='g', align='center')
plt.title('Random Data')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```
